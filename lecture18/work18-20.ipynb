```julia
using OrdinaryDiffEqTsit5
using PyPlot
```

```julia
using DataFrames
using CSV
```

```julia
function white_dwarf_eqs!(dudr, u, p, r)
    m = u[1]
    rho = u[2]
    rho_c = p
    if rho >= 0.0
        w = rho^(2/3)
        gamma = w/(3 * sqrt(1 + w))
        dudr[1] = rho * r * r
        if (r > 1.e-6)
            dudr[2] = -m * rho/(gamma * r * r)
        else
            dudr[2] = -rho_c/3 * r * rho/gamma
        end
    else
        dudr[1] = 0.0
        dudr[2] = 0.0
    end
end
```

```julia
rho_c1 = 0.08      #  min density in the star center
rho_c2 = 100000.0  #  max density in the star center
np = 100

rhos = exp.(range(log(rho_c1), log(rho_c2), np))
```

```julia
radi = zeros(np)
mass = zeros(np);
```

```julia
rspan = (0.0, 10.0)
```

```julia
for i = 1:np

    rho_c = rhos[i]
    u0 = [0.0, rho_c]
    prob = ODEProblem(white_dwarf_eqs!, u0, rspan, rho_c)

    condition(u, t, integrator) = u[2]
    affect!(integrator) = terminate!(integrator)
    cb = ContinuousCallback(condition, affect!)

    sol = solve(prob, Tsit5(), callback=cb)
    
    radi[i] = sol.t[end]
    mass[i] = sol.u[end][1]
end
```
